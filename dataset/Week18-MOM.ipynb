```python
year =[]
rank= []
c1 =[]
c2 = []


for k in range(391):
    for i in data1[data1.columns.unique()[0]]:
        year.append(i)
        
for k in range(391):
    for i1 in data1[data1.columns.unique()[1]]:
        c1.append(i1)
        
for k in range(391):
    for i2 in data1[data1.columns.unique()[2]]:
        c2.append(i2)
        rank.append(k)
        


       
        
```

```python
df = pd.DataFrame (year,columns=['Year'])
df["rank"] = rank
df["Realized CEO compensation"] = c1
df["Granted CEO compensation"] = c2
```

```python
decades= [1960,1970,1980,1990,2000,2010]
avg_c1 = [23.28,27,49.13,153.05,261.94,297.81]
avg_c2 = [17.0 , 19.8 , 36.0 , 154.1 ,253.4 ,208.3 ]
d1 = []
c1_=[]
c2_ =[]
rank1=[]

for k in range(391):
    for i in decades:
       d1.append(i)
        
for k in range(391):
    for i1 in avg_c1:
        c1_.append(i1)
        
for k in range(391):
    for i2 in avg_c2:
        c2_.append(i2)
        rank1.append(k)
```

```python
df1 = pd.DataFrame (d1,columns=['Decades'])
df1["rank1"] = rank1
df1["Avg. Realized CEO compensation"] = c1_
df1["Avg. Granted CEO compensation"] = c2_
```

```python
df1.to_csv("CEO_dataset2.csv")
```
